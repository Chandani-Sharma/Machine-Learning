## **5. Hypothesis Testing**

### Hypothetical Statement - 1
**Research Question:** Is there a significant difference in the energy consumption of appliances (Appliances) during weekdays and weekends?

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- Null Hypothesis (H0): There is no significant difference in energy consumption between weekdays and weekends.
- Alternative Hypothesis (HA): There is a significant difference in energy consumption between weekdays and weekends.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
#Extract the day of the week from the 'date' column in your DataFrame:
df['day_of_week'] = df['date'].dt.dayofweek

#Group the data by 'day_of_week' and calculate the mean energy consumption for each day:
energy_weekdays = df.loc[df['day_of_week'] < 5, 'Appliances']
energy_weekends = df.loc[df['day_of_week'] >= 5, 'Appliances']
mean_weekdays = energy_weekdays.mean()
mean_weekends = energy_weekends.mean()

#Perform the t-test and calculate the p-value:
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(energy_weekdays, energy_weekends)

print(t_stat, p_value)
```

##### Which statistical test have you done to obtain P-Value?

To obtain the p-value, a t-test was conducted. The t-test is a statistical test used to compare the means of two groups and determine if there is a significant difference between them. In this case, a t-test was performed to compare the energy consumption of appliances between weekdays and weekends. The t-test calculates the t-statistic, which measures the difference between the means of the two groups relative to the variability within each group. The p-value is then obtained from the t-statistic and represents the probability of observing the observed difference (or a more extreme difference) between the groups, assuming the null hypothesis is true.

##### Why did you choose the specific statistical test?

The t-test was chosen as the specific statistical test in this case because we are comparing the means of two groups (weekday and weekend) to determine if there is a significant difference in the energy consumption of appliances. The t-test is commonly used for such comparisons and is suitable when the data follows a normal distribution and the groups being compared are independent. In this case, we assume that the energy consumption on weekdays and weekends are independent of each other. Therefore, the t-test is an appropriate choice to assess the statistical significance of the difference in means between the two groups.

#### Insight from this hypothesis

- Based on the results of the t-test, the t-statistic value is approximately -2.4498 and the p-value is approximately 0.0143.

- Since the p-value (0.0143) is less than the chosen significance level (α = 0.05), we can reject the null hypothesis. This means that there is a significant difference in energy consumption between weekdays and weekends.

- In other words, the analysis suggests that the energy consumption of appliances differs significantly between weekdays and weekends.

### Hypothetical Statement - 2
**Research Question:** Is there a significant difference in the energy consumption of appliances (Appliances) during the day (9AM - 10PM) and during the night (10PM - 9AM)?

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- Null Hypothesis (H0): There is no significant difference in energy consumption between 2 PM - 9 PM and 10 PM - 1 PM.
- Alternative Hypothesis (HA): There is a significant difference in energy consumption between 2 PM - 9 PM and 10 PM - 1 PM.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy import stats

# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract the energy consumption values during the day and at night
day_energy = df.loc[(df['date'].dt.hour >= 14) & (df['date'].dt.hour <= 21), 'Appliances']
night_energy = df.loc[((df['date'].dt.hour >= 21) & (df['date'].dt.hour <= 23)) | ((df['date'].dt.hour >= 0) & (df['date'].dt.hour <= 2)), 'Appliances']

# Perform the two-sample independent t-test
t_statistic, p_value = stats.ttest_ind(day_energy, night_energy)

# Print the t-statistic and p-value
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
```

#### Insight from this hypothesis

- The obtained t-statistic value is 38.8193920204217, and the p-value is approximately 6.6576948325575e-310.

- The p-value is significantly smaller than the chosen significance level (α), suggesting strong evidence to reject the null hypothesis. Therefore, we can conclude that there is a significant difference in energy consumption between the day (9AM - 10PM) and night (10PM - 9AM) periods.
- Based on the results, we can say that energy consumption during day is much more than night.

##### Which statistical test have you done to obtain P-Value?

In this case, the statistical test used to obtain the p-value is the independent samples t-test. This test compares the means of two independent groups (day and night periods) to determine if there is a significant difference between them in terms of energy consumption.

##### Why did you choose the specific statistical test?

The independent samples t-test is chosen for this hypothesis test because it is appropriate when comparing the means of two independent groups. In this case, we are comparing the energy consumption during the day and night periods, which are distinct and independent groups. The t-test allows us to determine if there is a significant difference in energy consumption between these two periods based on their sample means.

## **7. ML Model Implementation**

### We will be looking at following Algorithms

1. Linear regression model

-  Improved Linear regression models

 2.  Ridge regression
 
 3.  Lasso regression
 
- Ensmble models

  4. ExtraTrees Regressor

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import ExtraTreesRegressor

# Define the models
linear_model = LinearRegression()
ridge_model = Ridge(alpha=0.5)
lasso_model = Lasso(alpha=0.1)

# Create a dictionary to store the models
models = [['Linear Regression:', linear_model], ['Ridge:', ridge_model], ['Lasso:', lasso_model],
           ['ExtraTreeRegressor :',ExtraTreesRegressor()]]

models_data = []
# Perform cross-validation for each model 
# Run all the proposed models and update the information in a list models_data
for model_name, model in models:
    model_data = {}
    model_data['Name'] = model_name
    model.fit(x_train, y_train)
    cv_scores = cross_val_score(model, independent_variables, dependent_variable, cv=5, scoring='neg_mean_squared_error')
    model_data['cv_scores'] = -cv_scores.mean()
    model_data['r2_scores'] = metrics.r2_score(y_test, model.predict(x_test))
    model_data['Mean Squared Error (MSE)'] = metrics.mean_squared_error( y_test, model.predict(x_test))
    model_data['sqrt (MSE)'] = np.sqrt(metrics.mean_squared_error( y_test, model.predict(x_test)))
    
   
    model_data['variance_score'] = model.score(x_test, y_test)
    
    models_data.append(model_data)
```
